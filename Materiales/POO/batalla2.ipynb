```python


import random
import time
from IPython.display import clear_output

# Superclase Guerrero
class Guerrero:
    def __init__(self, nombre: str):
        self.nombre = nombre
        self.salud = 0
        self.fuerza = 0
        self.defensa = 0
        self.velocidad = 0

    def atacar(self, objetivo):
        pass

    def defender(self, daño):
        pass

    def sigue_vivo(self):
        return self.salud > 0

    def mostrar_estadisticas(self):
        estatus = {
            'Nombre': self.nombre,
            'Fuerza': self.fuerza,
            'Defensa': self.defensa,
            'Velocidad': self.velocidad,
            'Salud': self.salud
        }
        return '\n'.join([f'{k}: {v}' for k, v in estatus.items()])

# Subclase Caballero
class Caballero(Guerrero):
    def __init__(self, nombre):
        super().__init__(nombre)
        self.salud = random.randint(100, 150)
        self.fuerza = random.randint(30, 40)
        self.defensa = random.randint(50, 70)
        self.velocidad = random.randint(10, 20)

    def atacar(self, objetivo):
        danio = self.fuerza * 1.2
        objetivo.defender(danio)

    def defender(self, danio):
        danio_recibido = danio - (self.defensa * 0.6)
        self.salud -= max(danio_recibido, 0)

# Subclase Arquero
class Arquero(Guerrero):
    def __init__(self, nombre):
        super().__init__(nombre)
        self.salud = random.randint(70, 100)
        self.fuerza = random.randint(25, 35)
        self.defensa = random.randint(20, 30)
        self.velocidad = random.randint(40, 60)

    def atacar(self, objetivo):
        if random.random() < 0.3:  # 30% probabilidad
            danio = int(self.fuerza * 1.0)
            objetivo.defensa = objetivo.defensa // 2  # Reduce defensa del objetivo
        else:
            danio = (self.fuerza * 1.0)
            objetivo.defender(danio)

    def defender(self, danio):
        if random.random() < 0.2:  # 20% probabilidad de esquivar
            return
        daño_recibido = danio - (self.defensa * 0.4)
        self.salud -= max(daño_recibido, 0)

# Subclase Mago
class Mago(Guerrero):
    def __init__(self, nombre):
        super().__init__(nombre)
        self.salud = random.randint(50, 80)
        self.fuerza = random.randint(40, 60)
        self.defensa = random.randint(10, 20)
        self.velocidad = random.randint(20, 30)

    def atacar(self, enemigo, todos_enemigos):
        if random.random() < 0.1:  # 10% probabilidad de ataque mágico
            for e in todos_enemigos:
                if e.sigue_vivo():
                    daño = int(self.fuerza * 0.8)  # daño a todos
                    e.defender(daño)
        else:
            daño = int(self.fuerza * 1.5)
            enemigo.defender(daño)

    def defender(self, daño):
        if random.random() < 0.1:  # 10% de esquivar
            return
        daño_recibido = daño - int(self.defensa * 0.4)
        self.salud -= max(daño_recibido, 0)

    def curar_objetivo(self, aliado):
        if aliado.sigue_vivo():
            curacion = min(15, 100 - aliado.salud)  # cura hasta 15 o lo que falta
            aliado.salud += curacion
            print(f'{self.nombre} ha curado a {aliado.nombre} por {curacion} puntos de salud.')

# Subclase Asesino
class Asesino(Guerrero):
    def __init__(self, nombre):
        super().__init__(nombre)
        self.salud = random.randint(60, 90)
        self.fuerza = random.randint(35, 45)
        self.defensa = random.randint(15, 25)
        self.velocidad = random.randint(50, 70)

    def atacar(self, objetivo):
        probabilidad = random.random() < 0.25  
        if probabilidad: 
            danio = self.fuerza * 2.0
        else:
            danio = self.fuerza * 1.0
        objetivo.defender(danio)

    def defender(self, danio):
        probabilidad = random.random() < 0.3  # 30% probabilidad de esquivar
        if probabilidad:
            return
        danio_recibido = danio - self.defensa * 0.15
        self.salud -= max(danio_recibido, 0)

# Subclase Curandero
class Curandero(Guerrero):
    def __init__(self, nombre):
        super().__init__(nombre)
        self.salud = random.randint(80, 120)
        self.fuerza = random.randint(10, 20)
        self.defensa = random.randint(20, 30)
        self.velocidad = random.randint(30, 40)

    def atacar(self, objetivo):
        daño = self.fuerza * 0.8
        objetivo.defender(daño)

    def curar(self, aliado):
        if aliado.sigue_vivo():
            curacion = min(15, 100 - aliado.salud)  # cura hasta 15 o lo que falta
            aliado.salud += curacion
            print(f'{self.nombre} ha curado a {aliado.nombre} por {curacion} puntos de salud.')

    def defender(self, daño):
        daño_recibido = daño - (self.defensa * 0.5)
        self.salud -= max(daño_recibido, 0)

# Función para crear un ejército de guerreros
def crear_ejercito(nombre_ejercito):
    return [
        Caballero(nombre=f"{nombre_ejercito} Caballero"),
        Arquero(nombre=f"{nombre_ejercito} Arquero"),
        Mago(nombre=f"{nombre_ejercito} Mago"),
        Asesino(nombre=f"{nombre_ejercito} Asesino"),
        Curandero(nombre=f"{nombre_ejercito} Curandero")
    ]

# Función para ejecutar la batalla
def batalla(ejercito1, ejercito2):
    
    while True:
        clear_output(wait=True)
        # Ordenar guerreros por velocidad
        turnos = sorted(ejercito1 + ejercito2, key=lambda x: x.velocidad, reverse=True)
        print(f'{turnos}')
        for guerrero in turnos:
            if guerrero.sigue_vivo():
                # Elegir un enemigo del ejército contrario
                if guerrero in ejercito1:
                    enemigos = [g for g in ejercito2 if g.sigue_vivo()]
                else:
                    enemigos = [g for g in ejercito1 if g.sigue_vivo()]

                if enemigos:  #  hay enemigos disponibles para atacar
                    enemigo = random.choice(enemigos)

                    # Decidir la acción dependiendo de la subclase
                    if isinstance(guerrero, Curandero):
                        # 50% probabilidad de curar a un aliado
                        if random.random() < 0.5:
                            aliados = [g for g in ejercito1 if g.sigue_vivo() and g != guerrero] if guerrero in ejercito1 else [g for g in ejercito2 if g.sigue_vivo() and g != guerrero]
                            if aliados:
                                aliado = random.choice(aliados)
                                guerrero.curar(aliado)  # Curar a un aliado
                                if aliado.sigue_vivo():
                                    print(f"{aliado.nombre} ahora tiene {aliado.salud} de salud.")
                                else:
                                    print(f"{aliado.nombre} fue curado, pero ya no sigue vivo.")
                            else:
                                guerrero.atacar(enemigo)  # Atacar si no hay aliados que curar
                        else:
                            guerrero.atacar(enemigo)
                    elif isinstance(guerrero, Mago):
                        guerrero.atacar(enemigo, ejercito1 + ejercito2)  # Pasa todos los enemigos
                    else:
                        guerrero.atacar(enemigo)

                    print(f"{guerrero.nombre} atacó a {enemigo.nombre}.")
                    print(guerrero.mostrar_estadisticas())
                    print(enemigo.mostrar_estadisticas())

                # Verificar si un ejército ha ganado
                if not any(g.sigue_vivo() for g in ejercito1):
                    print("El ejército 2 ganó la batalla!")
                    print("Juego terminado.")
                    return
                elif not any(g.sigue_vivo() for g in ejercito2):
                    print("El ejército 1 ganó la batalla!")
                    print("Juego terminado.")
                    return

        time.sleep(10)  # Pausa para ver cada turno
```
