```python
query = """
CREATE TABLE IF NOT EXISTS apple (
  id BIGINT,
  track_name VARCHAR(300),
  size_bytes BIGINT,
  price FLOAT(5,2),
  rating_count_tot BIGINT,
  rating_count_ver BIGINT,
  user_rating FLOAT(5,2),
  user_rating_ver FLOAT(5,2),
  prime_genre VARCHAR(100),
  PRIMARY KEY (id)
)
"""
```

```python
cursor.execute(query)
```

```python
cursor.execute("SHOW TABLES")
```

```python
for respuesta in cursor:
    print(respuesta)
```

```python
# VACIAR TABLA PORQUE ESTA DANDO ERROR POR CARACTERES
cursor.execute("TRUNCATE TABLE apple")
```

```python
for i, row in df.iterrows():
    query = """
INSERT INTO apple VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s)
    """
    # si se introduce la misma cantidad de valores no es necesario especificar las columnas
    values = (
        row['id'],
        row['track_name'],
        row['size_bytes'],
        row['price'],
        row['rating_count_tot'],
        row['rating_count_ver'],
        row['user_rating'],
        row['user_rating_ver'],
        row['prime_genre']
    )
    print(query, values)
    cursor.execute(query, values)
    cursor.execute("COMMIT") # para que guarde los cambios de lo contrario no se ejecutan
    (print(f"{i+1} filas insertadas"))
```
