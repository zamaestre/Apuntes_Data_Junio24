```python
# Escrapear informacion de libreria
lista_libreria=[]
lista_link_completo = []# Aqui tengo la lista de todos los links
lista_links= []
lista_nombre = []
lista_precio = []
lista_status_stock = []
lista_link_completo_imagen = []
lista_product_description = []
lista_product_tipo =[]
lista_Price_incl_tax = []
lista_Price_excl_tax = []
lista_tax = []
lista_Number_of_reviews = []
lista_UPC = []
lista_warning = []

for libro in lista_link_completo[0:3] : # Inicia el bucle a recorrer la lista de los links de los libros

    html3 = req.get(libro).content
    soup = bs(html3,'html.parser')
    
    
    dicc_libros = {} # creo mi diccionario para ir apendenado la informacion del libro
    

    # Columna Nombre
    nombre_libro = soup.find('li', {'class':'active'}).text # Busca el nombre del libro
    lista_nombre.append(nomre_libro)
    dicc_nombre['Nombre']= lista_nombre # apendea en el diccionario el nombre del libro
    dicc_nombre = {}
    print(f'{ nomre_libro}')

    # Link
    lista_links.append(libro)
    dicc_link['Link']= lista_links
    dicc_link = {}
    print(f'{libro}')


    # Columna Precio

    Precio = soup.find('p').text.replace('£',' ')
    lista_precio.append(Precio)
    dicc_precio['Precio']= lista_precio # apendea en el diccionario el nombre del libro
    dicc_precio = {}

    print(f'{ Precio }')

    # Columna Stock
    status_stock= soup.find('p', {'class' : 'instock availability'}).text.strip()
    lista_status_stock.append(status_stock)
    dicc_Status['Status Stock']= lista_status_stock
    dicc_Status= {}

    print(f'{ status_stock }')


    # Portada del libro
            
    link_completo_imagen =  completar_link + imagen_libro
    lista_link_completo_imagen.append(link_completo_imagen)
    dicc_portada['Portada']= lista_link_completo_imagen
    dicc_portada= {}



    print(f'{ link_completo_imagen }')

    # Descripcion del Libro

    product_description = soup.find('meta' , {'name' : 'description'}).attrs['content'].replace('\n', ' ').strip()
    lista_product_description.append(product_description)
    dicc_product_description['Descripcion']= lista_product_description
    dicc_product_description={}

    print(f'{product_description}')

    # Informacion de la tabla de la pagina del libro 
    Product_Type = soup.find_all('td')
            
    # UPC
    UPC= Product_Type[0].get_text(strip=True)
    lista_UPC.append(UPC)
    dicc_UPC[' UPC']=  lista_UPC
    dicc_UPC ={}
    print(f'{ UPC }')

    # Tipo de Producto
    tipo_Product =Product_Type[1].get_text(strip=True)
    lista_product_tipo.append(tipo_Product)
    dicc_tipo_Product[' Tipo de Producto']=  lista_product_tipo
    dicc_tipo_Product={}

    print(f'{Product }')

    #Precio con Impuesto
    Price_incl_tax = Product_Type[3].get_text(strip=True).replace('£', ' ')
    lista_Price_incl_tax.append(Price_incl_tax )
    dicc_Price_incl_tax[' Precio con Impuesto']= lista_Price_incl_tax
    dicc_Price_incl_tax ={}

    print(f'{Price_incl_tax }')

    # Precio sin Impuesto
    Price_excl_tax = Product_Type[2].get_text(strip=True).replace('£', ' ')
    lista_Price_excl_tax.append(Price_excl_tax)
    dicc_Price_excl_tax[' Precio sin Impuesto']= lista_Price_excl_tax
    dicc_Price_excl_tax ={}

    print(f'{Price_excl_tax}')

    # Impuestos
    Tax = Product_Type[4].get_text(strip=True).replace('£', ' ')
    lista_tax.append(Tax)
    dicc_Tax['Impuesto']= lista_tax
    dicc_Tax ={}
    print(f'{Tax}')

    # Reviws
    Number_of_reviews = Product_Type[6].get_text(strip=True)
    lista_Number_of_reviews.append(Number_of_reviews)
    dicc_Number_of_reviews['Nuemro de Reviews']= lista_Number_of_reviews 
    dicc_Number_of_reviews = {}
    print(f'{Number_of_reviews}')

    # warning 
    warning = soup.find('div', {'class': 'alert alert-warning'}).text.strip()
    lista_warning.append(warning)
    dicc_warning['Warning']= lista_warning
    dicc_warning= {}

    print(f'{warning}')

lista_libreria.append(dicc_nombre) # apendea el diccionario en la lista
lista_libreria.append(dicc_link) # apendea el diccionario en la lista
lista_libreria.append(dicc_precio) # apendea el diccionario en la lista
lista_libreria.append(dicc_Status) # apendea el diccionario en la lista
lista_libreria.append(dicc_portada) # apendea el diccionario en la lista
lista_libreria.append(dicc_product_description) # apendea el diccionario en la lista
lista_libreria.append(dicc_UPC) # apendea el diccionario en la lista
lista_libreria.append(dicc_tipo_Product) # apendea el diccionario en la lista
lista_libreria.append(dicc_Price_incl_tax ) # apendea el diccionario en la lista
lista_libreria.append(dicc_Price_excl_tax) # apendea el diccionario en la lista
lista_libreria.append(dicc_Tax) # apendea el diccionario en la lista
lista_libreria.append(dicc_Number_of_reviews) # apendea el diccionario en la lista
lista_libreria.append(dicc_warning) # apendea el diccionario en la lista

        
print('----------------------------------------------------------------------------------------------------')

   

# Aunque recorre el bucle solo guarda un registro en el dicc
    
       



```
